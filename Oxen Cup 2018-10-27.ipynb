```python
import numpy as np
import tolka_resultatlista as tolk
import plot_skatescores as plot
import matplotlib.pyplot as pl
pl.rcParams['figure.figsize'] = (12.0, 9.0)
```

```python
korta = tolk.tolka_TimeSchedule_resultat("results/tabula-Seniorer_A_Damer_Segment_1_JD_Mariestad-2018-10-27.csv")
langa = tolk.tolka_TimeSchedule_resultat("results/tabula-Seniorer_A_Damer_Segment_2_JDMariestad-2018-10-28.csv")
```

```python
k_pc = [el["PCS"] for el in korta]
k_el = [el["TES"] for el in korta]
k_name = [el["name"] for el in korta]
```

```python
plot.kosttnerplot(k_name,k_el,k_pc,"Senior A, Korta, Mariestad")
```

```python
l_pc = [el["PCS"] for el in langa]
l_el = [el["TES"] for el in langa]
l_name = [el["name"] for el in langa]
```

```python
plot.kosttnerplot(l_name,l_el,l_pc,"Senior A, Långa, Mariestad")
```

```python
goes = []
for skater in korta:
    for elem in skater["elements"]:
        goes.append(elem["GOE"])
k_goes = np.array(goes)

goes = []
for skater in langa:
    for elem in skater["elements"]:
        goes.append(elem["GOE"])
l_goes = np.array(goes)
```

```python
pl.hist(k_goes,13)
pl.title("Kortprogram")
pl.xlabel("GOE")
pl.ylabel("Frekvens")
pl.show()

pl.hist(l_goes,13)
pl.title("Långprogram")
pl.xlabel("GOE")
pl.ylabel("Frekvens")
pl.show()
```
